```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/MW-NIFTY-50-22-Mar-2024.csv")
df.head()
```

```python
for i in df.columns[1:]:
  df[i] = df[i].apply(lambda j :float(j.replace(',', "")))
```

```python
for column in df.columns[1:]:
    df[column] = df[column].apply(lambda x: float(x.replace(',', '')) if isinstance(x, str) else x)
```

```python
len(df)
```

```python
df.set_index("SYMBOL \n", inplace=True)
```

```python
wcss = []
for i in range(1,52):
  model = KMeans(i)
  model.fit(df)
  wcss.append(model.inertia_)
print(wcss)
```

```python
sns.set()
plt.plot(range(1,11), wcss[0:10])
plt.show()
```

```python
model = KMeans(4)
model.fit(df)

identify_cluster = model.fit_predict(df)
df['cluster']= identify_cluster

df.cluster.value_counts()
```
